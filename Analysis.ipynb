```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("build_stats.txt")
df
```

```python
#define index column
df.set_index('k', inplace=True)

#group data by product and display sales as line chart
df.groupby('Type')['Build Time'].plot(x='k', legend=True)
plt.title("Build Time vs. k (Grouped by Genome Size)")
```

```python
#group data by product and display sales as line chart
df.groupby('Type')['Output Size'].plot(x='k', legend=True)
plt.title("Overhead vs. k (Grouped by Genome Size)")
```

```python
df1 = pd.read_csv("query_stats.txt")
df1
```

```python
long = df1.loc[(df1['Genome Type'] == 'Long')]

#define index column
long.set_index('k', inplace=True)

#group data by product and display sales as line chart
long.groupby(['Algorithm', 'Query Type'])['Query Time'].plot(x='k', legend=True)
plt.title("Query Time vs. k for Long Genomes (Grouped by Algorithm and Query Size)")
```

```python
short = df1.loc[(df1['Genome Type'] == 'Half')]

#define index column
short.set_index('k', inplace=True)

#group data by product and display sales as line chart
short.groupby(['Algorithm', 'Query Type'])['Query Time'].plot(x='k', legend=True)
plt.title("Query Time vs. k for Short Genomes (Grouped by Algorithm and Query Size)")
```

```python
longq = df1.loc[(df1['Query Type'] == 'Long')]

#define index column
longq.set_index('k', inplace=True)

#group data by product and display sales as line chart
longq.groupby(['Algorithm', 'Genome Type'])['Query Time'].plot(x='k', legend=True)
plt.title("Build Time vs. k for Long Queries(Grouped by Algorithm)")
```

```python
shortq = df1.loc[(df1['Query Type'] == 'Half')]

#define index column
shortq.set_index('k', inplace=True)

#group data by product and display sales as line chart
shortq.groupby(['Algorithm', 'Genome Type'])['Query Time'].plot(x='k', legend=True)
plt.title("Build Time vs. k (Grouped by Genome Size)")
```
